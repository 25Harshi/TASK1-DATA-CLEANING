```python
import pandas as pd

# Load dataset
df = pd.read_csv('netflix_titles.csv')

# View basic info
df.info()  # Check column names and data types
df.head()  # Preview the first few rows
```

```python
print(df.isnull().sum())  # Count of missing values per column
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Not Available')
df['rating'] = df['rating'].fillna('Not Rated')
df.head()
```

```python
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])  # Fill with most common date
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')  # Clean column names
df['type'] = df['type'].str.lower().str.strip()
df['listed_in'] = df['listed_in'].str.lower().str.strip()
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

```

```python
print(df['date_added'].unique())  # See how dates are formatted
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
print(df['date_added'].isnull().sum())  # Check for NaT (missing values)
```

```python
df = df.sort_values(by='date_added', ascending=True)

```

```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['day_added'] = df['date_added'].dt.day
df['date_added']
```

```python
df['release_year'] = df['release_year'].astype(int)
```

```python
print(df['duration'].dtype)  # Show data type of duration
```

```python
df['duration'] = df['duration'].str.extract(r'(\d+)')  # Extract numbers
df['duration'] = pd.to_numeric(df['duration'])  # Convert to integer
```

```python
df['duration'] = df['duration'].astype(str).replace('nan', '')  # Convert NaN to empty strings
df['duration'] = df['duration'].str.extract(r'(\d+)')  # Extract numbers
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')  # Convert to numeric
```

```python
print(df.dtypes)  # Check if 'duration' is now an integer
```

```python
df.loc[:, 'duration'] = df['duration'].fillna(df['duration'].median())

```

```python
df['duration'] = df['duration'].fillna(df['duration'].median())
```

```python
print(df.info())  # Check column types and missing values
print(df.head())  # View first few rows
```

```python
column_order = ['show_id', 'type', 'title', 'director', 'cast', 'country',
                'date_added', 'release_year', 'rating', 'duration', 'listed_in', 'description']

df = df[column_order]  # Arrange columns in correct order
```

```python
df.to_csv('cleaned_netflix_titles.csv', index=False)
```
